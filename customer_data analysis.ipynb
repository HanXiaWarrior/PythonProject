```python
# calculate time and generate my output
def meetingtime(my_data, i):
    max_hour = int(my_data[i-2][2][0]+my_data[i-2][2][1])*3600
    min_hour = int(my_data[i-3][2][0]+my_data[i-3][2][1])*3600
    max_minute = int(my_data[i-2][2][3]+my_data[i-2][2][4])*60
    min_minute = int(my_data[i-3][2][3]+my_data[i-3][2][4])*60
    max_second = int(my_data[i-2][2][6]+my_data[i-2][2][7])
    min_second = int(my_data[i-3][2][6]+my_data[i-3][2][7])
    
    total_seconds = (max_hour-min_hour)+(max_minute-min_minute)+(max_second-min_second)
    return int(total_seconds)

def output_function(i,string):
    count  = 0.00
    i = 8
    total_meetings_7day = 0.00
    avg_metting_mins = 0.00
    avg_meeting_sec = 0.00
    country = '""'
    day = ''
    total_meeting_sec = 0
    while i <= num:
        if my_data[i][1]==string:
            country = my_data[i-5][1]+my_data[i-5][2]
            if my_data[i-3][1]!=day:
                day = my_data[i-3][1]
                total_meetings_7day = total_meetings_7day + 1.00
            temp = meetingtime(my_data, i)
            #print(temp)
            total_meeting_sec = total_meeting_sec + temp
            count = count+1
        i = i+10
    if country=='"UNITEDSTATES",':
        country = '"USA"'
    avg_meeting_sec = total_meeting_sec/count
    avg_metting_mins= avg_meeting_sec/60.00
    #print(country)
    print([['['],
 ['{'],
 ['"Company":', string],
 ['"total_meetings_7day":', total_meetings_7day],
 ['"avg_metting_mins":', avg_meeting_sec],
 ['"avg_meeting_sec":', avg_meeting_sec],
 ['"country":', country],
 ['},'], 
 [']']])
```
