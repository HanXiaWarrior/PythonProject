```python
from sklearn.feature_extraction.text import  CountVectorizer  

vectorizer = CountVectorizer()  
a = vectorizer.fit_transform(df['ProjectTitle'])

#b = [df['ProjectTitle'][1], df['ProjectTitle'][300], df['ProjectTitle'][500], df['ProjectTitle'][600], df['ProjectTitle'][700], df['ProjectTitle'][800], df['ProjectTitle'][1000], df['ProjectTitle'][1300], df['ProjectTitle'][1450], df['ProjectTitle'][1500], df['ProjectTitle'][1700], df['ProjectTitle'][1930], df['ProjectTitle'][2000], df['ProjectTitle'][2200]]
#print (temp['ProjectTitle'][2131])
b = df['ProjectTitle'][1]
bow =  vectorizer.fit_transform([b])
#print (bow)
#print (vectorizer.get_feature_names()[0])

from sklearn.naive_bayes import MultinomialNB 
spam_detect_model = MultinomialNB().fit(a,df['Sector'])

```

```python
all_predictions = spam_detect_model.predict(a)
from sklearn.metrics import classification_report
print (classification_report(df['Sector'], all_predictions))
```
